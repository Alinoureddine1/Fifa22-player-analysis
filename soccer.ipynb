```python
import pandas as pd
data_frame=pd.read_csv('data.csv')
data_frame
```

```python
df1=pd.DataFrame(data_frame,columns=['Name','WageEUR','ValueEUR'])
df1['Difference']=df1['ValueEUR']-df1['WageEUR']
df1.sort_values('Difference',ascending=False)
```

```python
import seaborn as sns
sns.set()
graph = sns.scatterplot(x='WageEUR', y='ValueEUR',data=df1)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(WageEUR,ValueEUR)", "(@WageEUR, @ValueEUR)"),
    ("Name", "@Name")])
    

p = figure(title="FIFA 22", x_axis_label='WageEUR', y_axis_label='ValueEUR',plot_width= 700, plot_height=700,tools=[TOOLTIPS])
p.circle('WageEUR','ValueEUR',size=10,source=df1)
show(p)
```
